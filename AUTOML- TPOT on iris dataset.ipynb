```python
import pandas as pd
import numpy as np
from tpot import TPOTClassifier  # classification
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
dir(iris)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=iris.data
y=iris.target
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

```python
y_test
```

```python
tc=TPOTClassifier(max_time_mins=10)
tc.fit(X_train,y_train)
print(tc.score(X_test,y_test))
```

```python
tc.evaluated_individuals_
```
